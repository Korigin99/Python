```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris.keys()
```

```python
display(iris.data, iris.target, iris.target_names, 'DESCR', 'feature_names', 'filename')
```

```python
plt.scatter(iris.data[:,0], iris.data[:,1], c=iris.target, alpha=0.3)
plt.colorbar()
```

```python
x = iris.data[:,2] #입력 데이터
y = iris.data[:,3] #목표값
```

```python
# X(입력값)은 항상 2차원 행렬이어야 한다.
X=iris.data[:,[2]] # X = iris.data[:,2].reshape(-1,1) 로 해도 됨
```

```python
X.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X) 
print(y_pred)
```

```python
plt.scatter(X[:,0],y)
```

```python
plt.scatter(X[:,0],y)
plt.plot(X[:,0], y_pred, 'r')
plt.axis('scaled')
```

```python
model.coef_ #weight 가중치 기울기
```

```python
model.intercept_ #bias b 절편
```

```python
model.predict([[1]]) # 2차원 형태
```

```python
model.predict([[7]])
```

```python
model.predict([[1],[7]])
```

```python
xs = np.array([1,7])
ys = model.coef_[0]*xs + model.intercept_
plt.scatter(X[:,0],y)
plt.plot(xs, ys, 'r')
```

```python
model.score(X,y)
```

```python
X = iris.data[:,:2]
y = np.array([0]*50 + [1]*100)
X.shape, y.shape 
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X,y)
y_pred = model.predict(X)
y_pred
```

```python
# 결과분석
model.coef_ # [w1, w2]
```

```python
model.intercept_ # b
```

```python
w = model.coef_[0]
b = model.intercept_[0]
w, b
```

```python
xs = np.array([4, 8])
ys = -(w[0]*xs + b)/w[1]

plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(xs, ys, 'r')
```

다중분류 (클래스가 3개 이상인 경우)

로지스틱회귀에서 기본적으로 2개의 클래스만 구분한다(0과 1)

다중분류일 경우 소프트맥스 함수를 적용한 일대다(ovr: one vs rest) 방식을 적용하다는 것이 일반적이다.

softmax 함수는 신경망에서 중요하게 다루어진다.

```python
X = iris.data[:,:2]
y = iris.target # 0, 1, 2  세 가지 품종을 분류
X.shape, y.shape
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
plt.colorbar()
```

```python
model = LogisticRegression()
model.fit(X,y)
y_pred = model.predict(X)
y_pred
```

```python
mglearn.plots.plot_2d_classification(model, X, fill=True, eps=0.5, alpha=0.4)
mglearn.discrete_scatter(X[:,0],X[:,1],y)
plt.legend(loc=4)
```
